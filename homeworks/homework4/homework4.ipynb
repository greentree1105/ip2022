```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = cv2.imread('dsu7.jpg')
img2 = cv2.imread('dsu8.jpg')
rows,cols,ch = img.shape

pts1 = np.float32([[259,290],[898,261],[259,354],[897,339]])
pts2 = np.float32([[370,301],[1138,300],[369,380],[1139,378]])
# ([[373,299],[1136,303],[372,374],[1137,379]])
M = cv2.getPerspectiveTransform(pts1,pts2)

dst = cv2.warpPerspective(img,M,(750,768))
output = img2[:,512:1024]

f_img1 = cv2.resize(dst,(512,768))
f_img2 = cv2.resize(output,(512,768))

addh = cv2.hconcat([f_img1,f_img2])
test = cv2.imwrite('test.jpg',addh)
plt.figure(figsize=(10,10))
plt.subplot(131),plt.imshow(img),plt.title('Input')
plt.subplot(132),plt.imshow(dst),plt.title('Output')
plt.subplot(133),plt.imshow(addh),plt.title('test')

 
plt.show()
```
