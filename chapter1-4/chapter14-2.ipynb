```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
#Scaling (늘리기)

img = cv2.imread('beach_r.jpg',0)

res = cv2.resize(img,None,fx=2, fy=2, interpolation = cv2.INTER_CUBIC)

#OR
height, width = img.shape[:2]
res = cv2.resize(img,(2*width, 2*height), interpolation = cv2.INTER_CUBIC)

img = cv2.imwrite('beach_rr.jpg', res)
```

```python
#Rotation (회전하기)

img = cv2.imread('beach_r.jpg',0)

rows,cols = img.shape

M = cv2.getRotationMatrix2D((cols/2,rows/2),90,1)
dst = cv2.warpAffine(img,M,(cols,rows))

```

```python
#affine Transformation (찌그러트리면서 돌리기?)

img = cv2.imread('camera.jpg')
rows,cols,ch = img.shape

pts1 = np.float32([[500,500],[2000,500],[500,2000]])
pts2 = np.float32([[100,1000],[2000,500],[1000,2500]])

M = cv2.getAffineTransform(pts1,pts2)

dst = cv2.warpAffine(img,M,(cols,rows))

plt.subplot(121),plt.imshow(img),plt.title('Input')
plt.subplot(122),plt.imshow(dst),plt.title('Output')
plt.show()
```

```python
img = cv2.imread('camera.jpg')
rows,cols,ch = img.shape

pts1 = np.float32([[153,117],[2645,229],[121,3721],[2609,3657]])
pts2 = np.float32([[0,0],[3000,0],[0,3000],[3000,3000]])

M = cv2.getPerspectiveTransform(pts1,pts2)

dst = cv2.warpPerspective(img,M,(3000,3000))

plt.subplot(121),plt.imshow(img),plt.title('Input')
plt.subplot(122),plt.imshow(dst),plt.title('Output')
plt.show()
```

```python
img = cv2.imread('lena.png')

kernel = np.ones((5,5),np.float32)/25
dst = cv2.filter2D(img,-1,kernel)

plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(dst),plt.title('Averaging')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
img = cv2.imread('lena.png')
blur = cv2.blur(img,(5,5))
plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(blur),plt.title('Blurred')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
import cv2
import numpy as np
img = cv2.imread('j.png',0)
img_o = cv2.imread('openj.png',0)
img_c = cv2.imread('closej.png',0)
kernel = np.ones((5,5),np.uint8)
erosion = cv2.erode(img,kernel,iterations = 1)
dilation = cv2.dilate(img,kernel,iterations = 1)
opening = cv2.morphologyEx(img_o, cv2.MORPH_OPEN, kernel)
closing = cv2.morphologyEx(img_c, cv2.MORPH_CLOSE, kernel)

```

```python
cv2.imshow('j',img)
cv2.imshow('erosion',erosion)
cv2.imshow('dilation',dilation)
cv2.imshow('openj',img_o)
cv2.imshow('opening',opening)
cv2.imshow('closej',img_c)
cv2.imshow('closing',closing)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
