```python
d['user']
```

```python
d['user'] = 'clown'
d
```

```python
d['id'] = 45
d
```

Removing dictionary entries


```python
d = {'user':'bozo', 'p':1234, 'i':34}

del d['user']
```

```python
d
```

```python
d.clear()
d
```

```python
a=[1,2]
del a[1]
a
```
